```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

from pandas.plotting import table
from pathlib import Path

pd.set_option('display.float_format', lambda x: '%.3f' % x)


# Apply consistent Matplotlib style
mpl.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times", "DejaVu Serif", "Computer Modern Roman"],
    "text.usetex": False,       # Set to True if you want LaTeX rendering
    "axes.labelsize": 16,
    "axes.titlesize": 17,
    "legend.fontsize": 14
})
```

# Experiment

```python
ID = ["ID"]
USER = ["SubjectID"]
IDS = ["SubjectID", "VideoID"]
TARGET = ["predefinedlabel"]
FEATURES = ["Delta", "Theta", "Alpha1", "Alpha2", "Beta1", "Beta2", "Gamma1", "Gamma2"]
```

```python
data_dir = Path("/home/aseliverstov/projects/brain_signals/data")
data = pd.read_csv(data_dir / "EEG_data.csv")

data["ID"] = (len(np.unique(data["VideoID"])) * data["SubjectID"] + data["VideoID"]).astype("int")
data = data[ID + USER + FEATURES + TARGET]

data.head(3)
```

```python
np.round(data[FEATURES].describe()).astype("int")
```

```python
k = 0
signals = ["Delta", "Theta", r"Beta1"]
feature_names = [r"$\delta$", r"$\theta$", r"$\beta_{1}$"]

plt.figure(figsize=(16, 6))  # same aspect ratio as in plot_metrics

cur_df = data[data["ID"] == k]
ts = np.arange(cur_df.shape[0])

# Use Matplotlib color cycle for consistency
color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

for i, _ in enumerate(signals):
    color = color_cycle[i % len(color_cycle)]
    plt.plot(ts, cur_df[signals].iloc[:, i],
             label=feature_names[i],
             linestyle='-',
             linewidth=1,
             color=color)

plt.yscale('log')
plt.xlabel("Timestep")
plt.ylabel("Feature Value")
# plt.title(f"Sample {k}. True target: {'Confusing' if cur_df[TARGET].iloc[0, 0].astype('int') else 'Not Confusing'}")
plt.title(f"Signal from sample 0 with not confusing label")
plt.grid(True)

# Same legend style: below plots in metrics version
plt.legend(loc='lower center', ncol=3, frameon=False, bbox_to_anchor=(0.5, -0.25))
plt.subplots_adjust(bottom=0.25)

plt.tight_layout()

plt.savefig("../../assets/signals_sample0.pdf", format="pdf", bbox_inches="tight", dpi=1200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times", "DejaVu Serif", "Computer Modern Roman"],
    "text.usetex": False,       # Set to True if you want LaTeX rendering
    "axes.labelsize": 16,
    "axes.titlesize": 17,
    "legend.fontsize": 14
})

label_map = {
    "Delta": r"$\delta$",
    "Theta": r"$\theta$",
    "Alpha1": r"$\alpha_{1}$",
    "Alpha2": r"$\alpha_{2}$",
    "Beta1": r"$\beta_{1}$",
    "Beta2": r"$\beta_{2}$",
    "Gamma1": r"$\gamma_{1}$",
    "Gamma2": r"$\gamma_{2}$",
    "predefinedlabel": r"$y$"
}


# Drop ID columns
data_clean = data.drop(columns=["ID", "SubjectID"], errors="ignore")

# Compute the correlation matrix
corr = data_clean.corr()

corr = corr.rename(index=label_map, columns=label_map)

# Create a mask for the upper triangle + diagonal
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap
sns.heatmap(
    corr,
    mask=mask,                  # remove diagonal & upper triangle
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",               # professional, symmetric colormap
    center=0,                    # center at zero correlation
    square=True,
    linewidths=0.5,
    linecolor="white",
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Matrix", fontsize=16, pad=12)
plt.tight_layout()

plt.savefig(
    "../../assets/corr_matrix.pdf",
    format="pdf",
    bbox_inches="tight",
    dpi=1200
)

plt.show()
```
