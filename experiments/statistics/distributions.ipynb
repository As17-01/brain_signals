```python
import numpy as np
import pandas as pd
import random

from pathlib import Path
from tqdm import tqdm
```

# Experiment

```python
ID = ["ID"]
USER = ["SubjectID"]
IDS = ["SubjectID", "VideoID"]
TARGET = ["predefinedlabel"]
FEATURES = ["Delta", "Theta", "Alpha1", "Alpha2", "Beta1", "Beta2", "Gamma1", "Gamma2"]
INIT_SEED = 6412
```

```python
data_dir = Path("/home/aseliverstov/projects/brain_signals/data")
data = pd.read_csv(data_dir / "EEG_data.csv")

data["ID"] = (len(np.unique(data["VideoID"])) * data["SubjectID"] + data["VideoID"]).astype("int")
data = data[ID + USER + FEATURES + TARGET]

data.head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(4, 2, figsize=(16, 20))
axes = axes.flatten()

for i, signal in enumerate(FEATURES):
    ax = axes[i]

    last_20 = data.groupby("ID").tail(20)
    top_20 = data.groupby("ID").head(20)

    sns.kdeplot(data=last_20[last_20["predefinedlabel"] == 1], x=signal, color="blue", label="Label 1 last 20", ax=ax, fill=True, common_norm=False)
    sns.kdeplot(data=top_20[top_20["predefinedlabel"] == 1], x=signal, color="green", label="Label 1 top 20", ax=ax, fill=True, common_norm=False)

    sns.kdeplot(data=last_20[last_20["predefinedlabel"] == 0], x=signal, color="red", label="Label 0 last 20", ax=ax, fill=True, common_norm=False)
    sns.kdeplot(data=top_20[top_20["predefinedlabel"] == 0], x=signal, color="yellow", label="Label 0 top 20", ax=ax, fill=True, common_norm=False)

    ax.set_title(f"Distribution of {signal}")
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(4, 2, figsize=(16, 20))
axes = axes.flatten()

for i, signal in enumerate(FEATURES):
    ax = axes[i]

    last_5 = data.groupby("ID").tail(5)
    top_5 = data.groupby("ID").head(5)

    sns.kdeplot(data=last_5[last_5["predefinedlabel"] == 1], x=signal, color="blue", label="Label 1 last 5", ax=ax, fill=True, common_norm=False)
    sns.kdeplot(data=top_5[top_5["predefinedlabel"] == 1], x=signal, color="green", label="Label 1 top 5", ax=ax, fill=True, common_norm=False)

    sns.kdeplot(data=last_5[last_5["predefinedlabel"] == 0], x=signal, color="red", label="Label 0 last 5", ax=ax, fill=True, common_norm=False)
    sns.kdeplot(data=top_5[top_5["predefinedlabel"] == 0], x=signal, color="yellow", label="Label 0 top 5", ax=ax, fill=True, common_norm=False)

    ax.set_title(f"Distribution of {signal}")
    ax.legend()

plt.tight_layout()
plt.show()
```
